```python
# import
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import os
```

```python
# Get data (only train data)
path = os.path.join('PreparedData', 'train.csv')
df = pd.read_csv(path)
```

```python
df.head(5)
```

```python
# Show all column names
df.columns
```

```python
# Feature Engineering (one hot encode)
df = pd.get_dummies(df, drop_first=True)
df
```

```python
# Test train split
X_train, X_test, y_train, y_test = train_test_split(df.drop('moved_after_2019', axis=1), df['moved_after_2019'], test_size=0.1, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, max_depth=12, random_state=40)
rf_model.fit(X_train, y_train)
```

```python
rf_model.score(X_test, y_test)
```

```python
# Load df_train
path = os.path.join('PreparedData', 'train.csv')
df_train = pd.read_csv(path)

# Feature Engineering (one hot encode)
df_train = pd.get_dummies(df_train, drop_first=True)
```

```python
# Do prediction on test data
path = os.path.join('PreparedData', 'test.csv')
df_test = pd.read_csv(path)

# Feature Engineering (one hot encode)
df_test = pd.get_dummies(df_test, drop_first=True)
```

```python
df_train
```

```python
rf_model = RandomForestClassifier(n_estimators=600, max_depth=12, random_state=40)

X = df_train.drop('moved_after_2019', axis=1).values
y = df_train['moved_after_2019'].values
rf_model.fit(X, y)
```

```python
# Predict
y_pred = rf_model.predict(df_test.values)
```

```python
y_pred
```

```python
# Add prediction to df_test
df_test['moved_after_2019'] = y_pred
```

```python
# Turn into submission by keeping only user id and moved_after_2019
df_test = df_test[['user_id', 'moved_after_2019']]
```
